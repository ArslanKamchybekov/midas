```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin

class DateFeatureTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.mean_year_ = None
        self.std_year_ = None
        
    def fit(self, X, y=None):
        # Calculate year statistics for normalization
        dates = pd.to_datetime(X['date'])
        self.mean_year_ = dates.dt.year.mean()
        self.std_year_ = dates.dt.year.std()
        return self
        
    def transform(self, X):
        X = X.copy()
        
        # Convert date to datetime
        X['date'] = pd.to_datetime(X['date'])
        
        # Day of week features
        X['day_of_week'] = X['date'].dt.dayofweek
        X['day_of_week_sin'] = np.sin(X['day_of_week'] * (2 * np.pi / 7))
        X['day_of_week_cos'] = np.cos(X['day_of_week'] * (2 * np.pi / 7))
        
        # Day of month features
        X['day_of_month'] = X['date'].dt.day
        X['day_of_month_sin'] = np.sin(X['day_of_month'] * (2 * np.pi / 31))
        X['day_of_month_cos'] = np.cos(X['day_of_month'] * (2 * np.pi / 31))
        
        # Month features
        X['month'] = X['date'].dt.month
        X['month_sin'] = np.sin(X['month'] * (2 * np.pi / 12))
        X['month_cos'] = np.cos(X['month'] * (2 * np.pi / 12))
        
        # Year features
        X['year_norm'] = (X['date'].dt.year - self.mean_year_) / self.std_year_
        X['year_sin'] = np.sin(X['year_norm'] * (2 * np.pi))
        X['year_cos'] = np.cos(X['year_norm'] * (2 * np.pi))
        
        return X
```

```python
class ProphetTransformer:
    def __init__(self, periods=30):
        self.periods = periods
        self.model = None
        self.forecast = None
        
    def fit(self, X, y=None):
        # Prepare data for Prophet
        prophet_df = X[['date', 'amount']].copy()
        prophet_df.columns = ['ds', 'y']  # Prophet requires these specific column names
        
        # Add circular features as regressors
        prophet_df['day_of_week_sin'] = X['day_of_week_sin']
        prophet_df['day_of_week_cos'] = X['day_of_week_cos']
        prophet_df['day_of_month_sin'] = X['day_of_month_sin']
        prophet_df['day_of_month_cos'] = X['day_of_month_cos']
        prophet_df['month_sin'] = X['month_sin']
        prophet_df['month_cos'] = X['month_cos']
        prophet_df['year_sin'] = X['year_sin']
        prophet_df['year_cos'] = X['year_cos']

        # Initialize and fit Prophet model
        self.model = Prophet()
        
        # Add regressors
        self.model.add_regressor('day_of_week_sin')
        self.model.add_regressor('day_of_week_cos')
        self.model.add_regressor('day_of_month_sin')
        self.model.add_regressor('day_of_month_cos')
        self.model.add_regressor('month_sin')
        self.model.add_regressor('month_cos')
        self.model.add_regressor('year_sin')
        self.model.add_regressor('year_cos')
        
        self.model.fit(prophet_df)
        return self
        
    def transform(self, X):
        # Create future dates dataframe
        future = self.model.make_future_dataframe(periods=self.periods)
        
        # Add circular features to future dataframe
        future_dates = pd.to_datetime(future['ds'])
        
        future['day_of_week'] = future_dates.dt.dayofweek
        future['day_of_week_sin'] = np.sin(future['day_of_week'] * (2 * np.pi / 7))
        future['day_of_week_cos'] = np.cos(future['day_of_week'] * (2 * np.pi / 7))
        
        future['day_of_month'] = future_dates.dt.day
        future['day_of_month_sin'] = np.sin(future['day_of_month'] * (2 * np.pi / 31))
        future['day_of_month_cos'] = np.cos(future['day_of_month'] * (2 * np.pi / 31))
        
        future['month'] = future_dates.dt.month
        future['month_sin'] = np.sin(future['month'] * (2 * np.pi / 12))
        future['month_cos'] = np.cos(future['month'] * (2 * np.pi / 12))
        
        future['year_norm'] = (future_dates.dt.year - X['year_norm'].mean()) / X['year_norm'].std()
        future['year_sin'] = np.sin(future['year_norm'] * (2 * np.pi))
        future['year_cos'] = np.cos(future['year_norm'] * (2 * np.pi))
        
        # Make predictions
        self.forecast = self.model.predict(future)
        return self.forecast
        
    def plot_forecast(self):
        if self.forecast is None:
            raise ValueError("Must call transform() before plotting")
        return self.model.plot(self.forecast)
        
    def plot_components(self):
        if self.forecast is None:
            raise ValueError("Must call transform() before plotting components")
        return self.model.plot_components(self.forecast)
```
